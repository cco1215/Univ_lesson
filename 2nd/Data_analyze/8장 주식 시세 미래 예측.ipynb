```python
import pandas as pd
import requests

new_url = 'https://finance.naver.com/item/sise_day.naver?code=252670&page='

my_headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
all_tables_year = pd.DataFrame()

for page_number in range(1, 100):
    full_url = new_url + str(page_number)
    
    print(f'{page_number}  .({full_url})')
    page = requests.get(full_url, headers=my_headers)
    
    table = pd.read_html(page.text)[0]
    print(f' {len(all_tables_year.index)}  {len(table.index)} .')
    all_tables_year = pd.concat([all_tables_year, table])
```

```python
all_tables_year.dropna(inplace=True)
all_tables_year.rename(columns={'': 'ds'}, inplace=True)
all_tables_year['y'] = all_tables_year[''] - ((all_tables_year[''] - all_tables_year['']) / 2)
all_tables_year
```

```python
df = pd.DataFrame()
df = all_tables_year[['ds', 'y']]
df.info()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=1000)

future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
import pandas as pd
import requests

new_url = 'https://finance.naver.com/item/sise_day.naver?code=039980&page='

my_headers = {'User-Agent': 'Mozilla/5.0'}
all_tables_year = pd.DataFrame()

for page_number in range(1, 100):
    full_url = new_url + str(page_number)
    page = requests.get(full_url, headers=my_headers)
    table = pd.read_html(page.text)[0]
    all_tables_year = pd.concat([all_tables_year, table])

all_tables_year.dropna(inplace=True)
all_tables_year.rename(columns={'': 'ds'}, inplace=True)
all_tables_year['y'] = all_tables_year[''] - ((all_tables_year[''] - all_tables_year['']) / 2)
all_tables_year

df = pd.DataFrame()
df = all_tables_year[['ds', 'y']]
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(periods=1000)

future.tail()
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
fig3 = m.plot(forecast)
```
