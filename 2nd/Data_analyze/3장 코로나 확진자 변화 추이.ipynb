```python
import pandas as pd
```

```python
rawdf = pd.read_csv('C:/jupiterr/owid-covid-data.csv')
```

```python
rawdf.info()
```

```python
select_colums = ['iso_code', 'location','date','total_cases','population']

reviseddf = rawdf[select_colums]

reviseddf.head()
```

```python
location = reviseddf['location']
location
```

```python
location.unique()
```

```python
kordf = reviseddf[reviseddf.location == 'South Korea']
kordf.head()
```

```python
kor_date_index_of = kordf.set_index('date')

kor_date_index_of.head()
```

```python
usadf = reviseddf[reviseddf.location == 'United States']

usa_date_index_of = usadf.set_index('date')

usa_date_index_of.head()
```

```python
kor_total_cases = kor_date_index_of['total_cases']
kor_total_cases
```

```python
usa_total_cases = usa_date_index_of['total_cases']
usa_total_cases
```

```python
kor_date_index_of.index
```

```python
finaldf = pd.concat([kor_date_index_of, usa_date_index_of], keys=["KOR", "USA"])

finaldf.head()
```

```python
finaldf['2022-01-01':].plot.line()
```

```python
kor_population = kor_date_index_of['population']['2022-01-01']

usa_population = usa_date_index_of['population']['2022-01-01']

print("(", usa_population, ") /  (", kor_population, ") = ", usa_population/kor_population)
```

```python
usa_rate = round((kor_population / usa_population), 2)
```

```python
final_revised_df = pd.DataFrame({'KOR' : kor_total_cases * usa_rate, 'USA' : usa_total_cases}, index=kor_date_index_of.index)

final_revised_df.head()
```

```python
final_revised_df[:].plot.line()
```

```python
final_revised_df['2022-01-01':].plot.line(rot=45)
```

```python
fra_df = reviseddf[reviseddf.location == 'France']

fra_date_index_df = fra_df.set_index('date')

fra_total_cases = fra_date_index_df['total_cases']

fra_population = fra_date_index_df['population']['2022-01-01']

fra_rate = round((kor_population / fra_population), 2)

final_revised_df = pd.DataFrame({'KOR' : kor_total_cases * fra_rate, 'FRA' : fra_total_cases}, index=kor_date_index_of.index)

final_revised_df[:].plot.line()
```

```python
gbr_df = reviseddf[reviseddf.location == 'United Kingdom']

gbr_date_index_df = gbr_df.set_index('date')

gbr_total_cases = gbr_date_index_df['total_cases']

gbr_population = gbr_date_index_df['population']['2022-01-01']

gbr_rate = round((kor_population / gbr_population), 2)
```

```python
pol_df = reviseddf[reviseddf.location == 'Poland']

pol_date_index_df = pol_df.set_index('date')

pol_total_cases = pol_date_index_df['total_cases']

pol_population = pol_date_index_df['population']['2022-01-01']

pol_rate = round((kor_population / pol_population), 2)

```

```python
final_revised_df = pd.DataFrame({'KOR' : kor_total_cases, 'USA' : usa_total_cases * usa_rate, 'FRA' : fra_total_cases * fra_rate, 'GBR' : gbr_total_cases * gbr_rate, 'POL' : pol_total_cases * pol_rate}, index=kor_date_index_of.index)
final_revised_df['2022-01-01':].plot.line(rot=45)
```
