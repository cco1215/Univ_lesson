```python
import pandas as pd
data = pd.read_csv('C:/jupiterr/archive/seoul-metro-2021.logs.csv')
data
```

```python
station_info = pd.read_csv('C:/jupiterr/archive/seoul-metro-station-info.csv')
station_info
```

```python
data.info()
```

```python
station_sum = data.groupby("station_code").sum()
station_sum
```

```python
station_info.info()
```

```python
station_info = station_info[['station.code', 'geo.latitude', 'geo.longitude']]
station_info
```

```python
station_info = station_info.set_index('station.code')
station_info
```

```python
joined_data = station_sum.join(station_info)
joined_data
```

```python
seoul_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
seoul_in
```

```python
from folium.plugins import HeatMap
HeatMap(data = joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(seoul_in)
seoul_in
```

```python
seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(seoul_out)
seoul_out
```

```python
data.info()
```

```python
morning_data = data[pd.to_datetime(data.timestamp).dt.hour < 9]
morning_data
```

```python
evenig_data = data[pd.to_datetime(data.timestamp).dt.hour > 17]
evenig_data
```

```python
morning_data_sum = morning_data.groupby("station_code").sum()
evenig_data_sum = evenig_data.groupby("station_code").sum()
```

```python
morning_joined_data = morning_data_sum.join(station_info)
evenig_joined_data = evenig_data_sum.join(station_info)
```

```python
morning_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(morning_seoul_in)
morning_seoul_in
```

```python
morning_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(morning_seoul_out)
morning_seoul_out
```

```python
evenig_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = evenig_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(evenig_seoul_in)
evenig_seoul_in
```

```python
evening_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = evenig_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(evening_seoul_out)
evening_seoul_out
```

```python
lunch_data = data[(pd.to_datetime(data.timestamp).dt.hour > 11)&(pd.to_datetime(data.timestamp).dt.hour < 14)]
lunch_data_sum = lunch_data.groupby("station_code").sum()
lunch_joined_data = lunch_data_sum.join(station_info)
```

```python
lunch_data_sum = lunch_data.groupby("station_code").sum()
lunch_data_sum
lunch_joined_data = lunch_data_sum.join(station_info)
```

```python
lunch_seoul_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = lunch_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(lunch_seoul_in)
lunch_seoul_in
```

```python
lunch_seoul_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
HeatMap(data = lunch_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(lunch_seoul_out)
lunch_seoul_out
```

```python
lunch_seoul_out_circle = folium.Map(location=[37.55, 126.98], zoom_start=12)
for index, row in lunch_joined_data[['geo.latitude', 'geo.longitude', 'people_out']].iterrows():
    if int(row['people_out']) > 2_500_00:
        folium.Circle(location=[str(row['geo.latitude']), str(row['geo.longitude'])], color='red', fill = True,
                      radius=(int(row['people_out'])/10000)).add_to(lunch_seoul_out_circle)
lunch_seoul_out_circle
```
